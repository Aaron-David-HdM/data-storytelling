```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="ticks", color_codes=True)
my_blue = "#3F83F4"
```

```python
# Import data
df = pd.read_excel('~/Downloads/2_5_data.xlsx')  
df
```

## Line plot

```python
sns.lineplot(data=df, 
             x="Year", 
             y="Rate")

sns.despine();
```

```python
sns.lineplot(data=df, 
            x="Year", 
            y="Rate")

plt.axhline(y=df.Rate.mean(), 
            color=my_blue,
            ls='--', 
            lw=0.5);

# Label and anotation

plt.text(2018, 0.07, 
        "AVG 7.5", 
        horizontalalignment='left', 
        size='medium', 
        color= my_blue, 
        weight='semibold')

sns.despine();
```

```python
sns.lineplot(data=df, 
            x="Year", 
            y="Rate")

plt.axhline(y=df.Rate.mean(), 
            color=my_blue,
            ls='--', 
            lw=0.5);

plt.fill_between(df.Year, df.Rate)

sns.despine();
```

```python
sns.lineplot(data=df, 
            x="Year", 
            y="Rate");

plt.axhline(y=df.Rate.mean(), 
            color=my_blue,
            ls='--', 
            lw=0.5);

plt.fill_between(df.Year, df.Rate.mean(), alpha=0.20)

sns.despine();
```

```python
sns.barplot(data=df, 
            x="Year", 
            y="Rate",
            color=my_blue);

plt.axhline(y=df.Rate.mean(), 
            color=my_blue,
            ls='--', 
            lw=0.5)

sns.despine();
```
